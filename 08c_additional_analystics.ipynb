```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv("05_d_m_additional_analystics.csv")

features = ['Score_with_stats', 'ast_per_game', 'stl_per_game', 'tov_per_game']
target = 'Sentiment Score'

corr = df[[target] + features].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("📊 Correlation Matrix: Sentiment vs Stats")
plt.tight_layout()
plt.show()

X = df[features]
y = df[target]

X = sm.add_constant(X)  
model = sm.OLS(y, X).fit()
print(model.summary())
```
